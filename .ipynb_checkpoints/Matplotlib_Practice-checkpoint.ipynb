```python
#importing matplotlib
import matplotlib.pyplot as plt
```

```python
#Making a simple line plot
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot
plt.plot(x, y)

# Labels aur Title
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

```python
#for making a bar chart
# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

# Bar Chart
plt.bar(categories, values)

# Labels aur Title
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()
```

```python
#scatter plot
# Data
x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 55]

# Scatter Plot
plt.scatter(x, y, color='red')

# Labels aur Title
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

```python
#pie chart
# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 35, 15, 10]

# Pie Chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Title
plt.title('Programming Language Popularity')

plt.show()
```

```python
#Making multiple plots in a single figure
# Data
x = [1, 2, 3, 4, 5]

# Subplots
plt.subplot(1, 2, 1)  # 1 row, 2 columns, Plot 1
plt.plot(x, [i**2 for i in x], label='Square')
plt.title('Square Function')

plt.subplot(1, 2, 2)  # 1 row, 2 columns, Plot 2
plt.plot(x, [i**3 for i in x], label='Cube', color='green')
plt.title('Cube Function')

plt.tight_layout()
plt.show()
```
