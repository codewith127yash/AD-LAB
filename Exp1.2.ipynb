```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#.Load a sample dataset
iris = sns.load_dataset('iris')
print("Sample of the iris dataset:")
print(iris.head())
```

```python
plt.figure(figsize=(8,6))
plt.hist(iris['sepal_length'], bins=10 ,color='pink', edgecolor='blue')
plt.title('Distrbution of petal length')
plt.xlabel('sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
#3. Create scatter plots to understand
# Sactter plot for 'sepalLength
plt.figure(figsize=(8,6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)

plt.title('scatter plot of sepal length vs petal width')
plt.xlabel('sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
#4. use a correlation heatmap to find relationships between multiple features

#Compute the coreelation matrix
correlation_matrix = iris.drop(columns=['species']).corr()
```

```python
# Plot the heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
