```python
from pandas import read_csv
from numpy import set_printoptions
from sklearn import preprocessing
data = pd.read_excel("CGPA_Salary_Data.xlsx")
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
data_scaler = StandardScaler().fit(data)
data_rescaled = data_scaler.transform(data)
data_rescaled
```

# Training and Testing
 x as cgpa (independent variable) , 
 y as salary (dependent variable)

```python
x = data[["CGPA"]]
y = data[["Salary (in LPA)"]]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from scipy.stats import zscore
import numpy as np
x = x[(np.abs(zscore(x)) < 3).all(axis=1)]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
params = {'alpha': [0.1, 1, 10, 100]}
ridge = GridSearchCV(Ridge(), param_grid=params, cv=5)
ridge.fit(x_train, y_train)
print(ridge.best_params_)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(x.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=100)
ridge_model.fit(x_train, y_train)
y_pred = ridge_model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("MSE:", mse)
print("R-squared:", r2)
```

```python
import matplotlib.pyplot as plt

residuals = y_test - y_pred
plt.scatter(y_pred, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.title("Residual Analysis with Ridge Regression")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.show()
```
