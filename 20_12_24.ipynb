# Training and Testing Data

```python
X = data[['CGPA']]
y = data[['Salary (in LPA)']]
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.3 , random_state = 10)
print(X_train)
```

# Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the data is already loaded
data = pd.read_excel("CGPA_Salary_Data.xlsx")

# Scatter plot with CGPA on the x-axis and Salary on the y-axis
plt.scatter(data['CGPA'], data['Salary (in LPA)'], color='blue', alpha=0.7)

# Adding labels and title
plt.xlabel("CGPA")
plt.ylabel("Salary (in LPA)")
plt.title("Scatter Plot of CGPA vs Salary")

# Show the plot
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
reg.predict([[7.98]])
```

```python
reg.predict(X_test)
```

```python
# Assuming reg is the trained LinearRegression model
input_cgpa = 7.98

# Extract scalar values from coef_ and intercept_
space = reg.coef_.item()  # Convert NumPy array to scalar
intercept = reg.intercept_.item()  # Convert NumPy array to scalar

# Calculate the prediction
predicted_salary = space * input_cgpa + intercept

# Display the result
print(f"Predicted Salary: {predicted_salary:.2f}")
```

# Model Evaluation

```python
from sklearn.metrics import mean_squared_error, r2_score
y_pred = reg.predict(X_test)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("Mean Squared Error = ",mse)
print("R-squared Error = " , r2)
```

# Visualization

```python
import matplotlib.pyplot as plt

# Assuming the model is already trained
# Generate predictions for the CGPA values using the trained model
y_pred = reg.predict(data[['CGPA']])

# Scatter plot of actual data points
plt.scatter(data['CGPA'], data['Salary (in LPA)'], color='blue', label='Actual Data')

# Plot the regression line using predicted values
plt.plot(data['CGPA'], y_pred, color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('CGPA')
plt.ylabel('Salary (in LPA)')
plt.legend()

# Show the plot
plt.show()
```
