```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Salary_Data.csv')
```

```python
df
```

```python
df.head(10)
```

```python
imputer = SimpleImputer(strategy='mean')
df[['Age', 'Years of Experience', 'Salary']]= imputer.fit_transform(df[['Age', 'Years of Experience', 'Salary']])
```

```python
df
```

```python
label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df['Education Level'] = label_encoder.fit_transform(df['Education Level'])
df['Job Title'] = label_encoder.fit_transform(df['Job Title'])
```

```python
df
```

```python
scaler = StandardScaler()
df[['Age', 'Years of Experience', 'Salary']] = scaler.fit_transform(df[['Age', 'Years of Experience', 'Salary']])
print("\nDataFrame after feature scaling :")
print (df)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Years of Experience', y='Salary', hue='Gender', data=df)

plt.title('scatter plot of Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
